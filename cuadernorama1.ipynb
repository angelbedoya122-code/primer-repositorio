```python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42) # You can use any integer value here. For quiz put 42

# Generate 25 random names for the 'nombre' column
nombres = [f'Estudiante{i}' for i in range(10)]

# Generate 4 columns of random grades between 0 and 5 with two decimal places
# np.random.uniform generates floats between low (inclusive) and high (exclusive)
parciales = np.round(np.random.uniform(0, 5, size=(10, 4)), 2);

# Create the DataFrame
df = pd.DataFrame({
    'nombre': nombres,
    'parcial1': parciales[:, 0],
    'parcial2': parciales[:, 1],
    'parcial3': parciales[:, 2],
    'parcial4': parciales[:, 3]
})

df
```

```python
df['Notafinal'] = 0.2*df['parcial1'] + 0.2*df['parcial2'] + 0.2*df['parcial3'] + 0.4*df['parcial4']
df
```

```python
def aprobado(dataframe):
    dataframe['Notafinal'] = 0.2*dataframe['parcial1'] + 0.2*dataframe['parcial2'] + 0.2*dataframe['parcial3'] + 0.4*dataframe['parcial4'] 
    cont = 0
    for i in range(df.shape[0]):
        if dataframe.iloc[i,5] >= 3.75:
            cont += 1
    return cont
```

```python
aprobado(df)
```
